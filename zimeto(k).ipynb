```python
df_country=pd.read_excel(r'Downloads\Country-Code.xlsx')
df_country.head(3)
```

```python
final_df=pd.merge(df, df_country, on='Country Code',how='left')
final_df.head(3)
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
counrty_val=final_df.Country.value_counts().values
counrty_val
```

```python
#the most transactions related to India
plt.pie(counrty_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'])
```

```python
rating=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
rating
```

```python
import seaborn as sns
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count',data=rating)
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color',data=rating,palette=['white',')
```

```python
sns.countplot(x='Rating color', data=rating, palette=['blue','red','orange','yellow','Green','darkGreen'])
```

```python
# Filter the DataFrame where 'Aggregate rating' is 0
filtered_df = final_df[final_df['Aggregate rating'] == 0]

# Group by 'Country' column
grouped_df = filtered_df.groupby('Country').size()
grouped_df
```

```python
# Convert the Series to a DataFrame and reset index
grouped_df = grouped_df.reset_index(name='Count')

# Display the DataFrame
print(grouped_df)
```

```python
# Selecting columns 'Country' and 'Currency' from final_df
selected_columns = final_df[['Country','Currency']]

# Grouping by 'Country' and 'Currency' and calculating the size of each group
grouped_df_currency = selected_columns.groupby(['Country','Currency']).size().reset_index()

# Displaying the resulting DataFrame
print(grouped_df_currency)
```

```python
grouped_df_currency
```

```python
# Selecting 'Country' and 'Has Online delivery' columns
selected_columns = final_df[['Country', 'Has Online delivery']]

# Filtering the DataFrame where 'Has Online delivery' is 'Yes'
filtered_df = selected_columns[selected_columns['Has Online delivery'] == 'Yes']

# Grouping by both 'Country' and 'Has Online delivery' columns and counting the occurrences
grouped_df = filtered_df.groupby(['Country', 'Has Online delivery']).size().reset_index(name='Count')

# Displaying the resulting grouped DataFrame
print(grouped_df)
```
