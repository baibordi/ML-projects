```python
import pandas as pd
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from collections import Counter

# Download NLTK data files (only need to run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()



# Helper function to get the part of speech tag for lemmatization
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

# Preprocess the text data
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove punctuation
    tokens = [word for word in tokens if word.isalnum()]
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    # Lemmatize tokens
    tokens = [lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in tokens]
    return tokens

# Apply preprocessing to the text column
merged_df['tokens'] = merged_df['text'].apply(preprocess_text)

# Verify the tokenization and lemmatization
print(merged_df[['text', 'tokens']])

# Group by age and calculate keyword frequencies
age_groups = merged_df.groupby('age')

# Create a dictionary to hold the Counter objects for each age group
age_group_keywords = {}

for age, group in age_groups:
    all_tokens = [token for tokens in group['tokens'] for token in tokens]
    keyword_counts = Counter(all_tokens)
    age_group_keywords[age] = keyword_counts

# Display common keywords for each age group
for age, keywords in age_group_keywords.items():
    print(f"Age group: {age}")
    print("Common keywords:", keywords.most_common(5))  # Display top 5 keywords
    print()
```
