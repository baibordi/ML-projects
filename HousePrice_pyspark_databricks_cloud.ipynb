This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

```
# File location and type
file_location = "/FileStore/tables/Housing-1.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "false"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
#First: categorical columns are by using the datatype as string
# Second: StringIndexer transformer is defined for columns with the previous name, just _index is added to the name of the new columns
# Third: Fit method is used to fit the transformer to the dataframe
from pyspark.ml.feature import StringIndexer

# Identify categorical columns
categorical_cols = [col for col in df.columns if df.select(col).dtypes[0][1] == 'string']

# Define StringIndexer for categorical columns
indexers = [StringIndexer(inputCol=col, outputCol=col + "_indexed") for col in categorical_cols]

# Apply StringIndexer to transform all categorical columns
indexed_df = df
for indexer in indexers:
    indexed_df = indexer.fit(indexed_df).transform(indexed_df)

# Show the transformed DataFrame
indexed_df.show()
```

```
# It is assumed that there is a dataframe containing columns that is transformed form the previous section from categorical to numerical indicies 
# OneHotEncoded is used to transform the data with indexies to new columns as an output

from pyspark.ml.feature import OneHotEncoder

# Assuming indexed_df is your DataFrame containing the indexed columns
indexed_cols = [col for col in indexed_df.columns if col.endswith("_indexed")]

# Define OneHotEncoder for indexed columns
encoder = OneHotEncoder(inputCols=indexed_cols, outputCols=[col.replace("_indexed", "_encoded") for col in indexed_cols])

# Apply OneHotEncoder to transform indexed columns
encoded_df = encoder.fit(indexed_df).transform(indexed_df)

# Show the transformed DataFrame
encoded_df.show()
```

```
#Combining both in a pipeline allows for a more streamlined and efficient transformation process, ensuring consistency and ease of usefrom pyspark.sql import SparkSession.

#StringIndexer converts categorical strings to numerical indices.
#OneHotEncoder converts numerical indices to one-hot encoded vectors.

from pyspark.ml.feature import StringIndexer, OneHotEncoder
from pyspark.ml import Pipeline

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Data Transformation") \
    .getOrCreate()

# Assuming df is your DataFrame
# Identify categorical columns
categorical_cols = [col for col in df.columns if df.select(col).dtypes[0][1] == 'string']

# Define StringIndexer for categorical columns
indexers = [StringIndexer(inputCol=col, outputCol=col + "_indexed") for col in categorical_cols]

# Define OneHotEncoder for indexed categorical columns
encoders = [OneHotEncoder(inputCol=col + "_indexed", outputCol=col + "_encoded") for col in categorical_cols]

# Assemble all indexers and encoders into a single pipeline
pipeline = Pipeline(stages=indexers + encoders)

# Fit the pipeline to the data
pipeline_model = pipeline.fit(df)

# Transform the DataFrame
transformed_df = pipeline_model.transform(df)

# Show the transformed DataFrame
transformed_df.show()
```

```
# A vector Assembler is used for merging and combing multiple columns to a single column whcih is named as features. In this way data is preperd for implementing a machine learning algotithm.

# fist I put all the columns in features which is wrong for implementing a linear regression because the target variable is seprate
from pyspark.ml.feature import VectorAssembler

# Assuming transformed_df is your DataFrame containing the encoded columns
# Identify the encoded columns
encoded_cols = [col for col in transformed_df.columns if col.endswith("_encoded")]

# Initialize VectorAssembler
assembler = VectorAssembler(inputCols=encoded_cols, outputCol="features")

# Apply VectorAssembler to the transformed DataFrame
final_df = assembler.transform(transformed_df)

# Show the final DataFrame
final_df.show()
```

```
# As a description for biginners i bring Vector assmbler again in this part for showing that when I put the features in vector assmbler, i should not put the target varibale. Atthe first step i put target varibale hich I encounterd error. So, In this section< I eliminated the target varible.

from pyspark.ml.feature import VectorAssembler

# Assuming transformed_df is your DataFrame containing the encoded columns
# Identify the encoded columns excluding _c0_encoded
encoded_cols = [col for col in transformed_df.columns if col.endswith("_encoded") and col != "_c0_encoded"]

# Initialize VectorAssembler
assembler = VectorAssembler(inputCols=encoded_cols, outputCol="features")

# Apply VectorAssembler to the transformed DataFrame
final_df = assembler.transform(transformed_df)

# Show the final DataFrame
final_df.show()
```

```
# First:The columns can be choosen in this part for implemeting a machine learning algotithm
#Second: The columns are transformed to a single column in order to be used in Linear Regression
#Third: Sepration between train and test dataset is a common step for supervised learning which is used as the next step
# Forth: Fit method is used for learning the parametrs of the model based on the features 
#Fifth: The train Regression model is used for prediction on test data
# Sixth: The lower the RMSE the better the accuracy we have, as the predicton calues are near to actual values

from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Linear Regression") \
    .getOrCreate()

# Assuming transformed_df is your DataFrame
# Identify the encoded columns excluding _c0_encoded
encoded_cols = [col for col in transformed_df.columns if col.endswith("_encoded") and col != "_c0_encoded"]

# Initialize VectorAssembler
assembler = VectorAssembler(inputCols=encoded_cols, outputCol="features")

# Apply VectorAssembler to the transformed DataFrame
final_df = assembler.transform(transformed_df)

# Show the final DataFrame
final_df.show()

# Split the data into training and testing sets
train_data, test_data = final_df.randomSplit([0.8, 0.2], seed=123)

# Initialize Linear Regression model
lr = LinearRegression(featuresCol="features", labelCol="_c12_indexed")

# Fit the model to the training data
lr_model = lr.fit(train_data)

# Make predictions on the test data
predictions = lr_model.transform(test_data)

# Show some sample predictions
predictions.select("_c12_indexed", "prediction").show()

# Evaluate the model
evaluator = RegressionEvaluator(labelCol="_c12_indexed", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE):", rmse)
```
